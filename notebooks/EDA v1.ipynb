```python
def show_metric_details(metric_name):
    if metric_name not in data.columns:
        print(metric_name, " not in columns")
        return()

    cnts = data[metric_name].value_counts()
    pcts = data[metric_name].value_counts(normalize=True)
    mets = pd.concat([cnts, pcts], keys=['Counts', 'Pcts'], axis=1)
    return(mets)
```
